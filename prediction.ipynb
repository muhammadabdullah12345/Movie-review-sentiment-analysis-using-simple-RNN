```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python
word_index=imdb.get_word_index()
reverse_index={word:key for key,word in word_index.items()}
```

```python
model=load_model('rnn_model.h5')
model.summary()
```

```python
# Step 2: Helper Functions
# Function to decode reviews
def decode_review(encoded_review):
    return ' '.join([reverse_index.get(i - 3, '?') for i in encoded_review])

# Function to preprocess user input
def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
def predict_review(review):
    preprocessed_review=preprocess_text(review)
    prediction=model.predict(preprocessed_review)
    predicted_review='Positive' if prediction[0][0] > 0.5 else 'Negative'
    return predicted_review,prediction[0][0]
```

```python
# example prediction
example_review = "This movie was extremely very fantastic! The acting was very much great and the plot was very much thrilling."

sentiment,score=predict_review(example_review)

print(f'Review: {example_review}')
print(f'Sentiment: {sentiment}')
print(f'Prediction Score: {score}')
```
